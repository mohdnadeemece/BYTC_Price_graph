```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Get data from Yahoo Finance
btc_data = yf.download('BTC-USD', start='2020-01-01', end='2024-12-31')

# View first few rows
btc_data.head()
```

```python
excel_filename = "my_data.xlsx"
btc_data.to_excel(excel_filename, index=False)
```

```python
df = pd.read_excel("/content/crypto_Btc_price.xlsx")
df
```

```python
plt.title("BTC_USD price across the years")
plt.plot(df["Date"], df["Price"])
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
# Read the Excel file again, skipping the first two rows and naming the columns
df = pd.read_excel("/content/crypto_Btc_price.xlsx", skiprows=2, names=["Date", "Price", "Close", "High", "Low", "Volume"])

# Convert the 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Display the first few rows to verify
display(df.head())
```
